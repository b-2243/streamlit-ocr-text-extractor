```python
import pytesseract
from pdf2image import convert_from_path
import os

# Set path to Tesseract executable
# Tesseract : Extracts text from images

pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

# Path to poppler bin folder
# Poppler : Converts PDF pages into images for OCR

poppler_path = r'C:\Users\BHAUMIK\Downloads\Release-24.08.0-0\poppler-24.08.0\Library\bin'  

# Convert PDF to images (each page becomes an image)
pdf_path = 'Patelbhaumik_resume.pdf'
pages = convert_from_path(pdf_path, dpi=300, poppler_path=poppler_path)  # DPI : Improves text detection accuracy

os.makedirs("output_pages", exist_ok=True)

# Loop over each page
for i, page_image in enumerate(pages):
    # Save page image (optional)
    image_path = f"output_pages/page_{i + 1}.png"
    page_image.save(image_path, 'PNG')

    # Extract text using OCR
    text = pytesseract.image_to_string(page_image)

    print(f"\n--- Page {i + 1} ---\n")
    print(text)
```
